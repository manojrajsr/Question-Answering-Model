```python
import random
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import warnings
import pandas as pd
warnings.filterwarnings('ignore')
```

```python
nltk.download('punkt', quiet=True)
```

```python
df = pd.read_csv("./Uttarakhand.csv")
df.columns = ["tweets"]
```

```python
import re
import gensim
from gensim.parsing.preprocessing import remove_stopwords


def clean_sentences(sentence , stopwords = False):
    sentence = sentence.lower().strip()
    sentence = re.sub(r'[^a-z0-9\s]', ' ', sentence)

    if stopwords:
        sentence = remove_stopwords(sentence)

    return sentence


def get_cleaned_sentences(df, stopwords = False):
    sents = df[["tweets"]]
    cleaned_sentences = []

    for index, row in df.iterrows():
        cleaned = clean_sentences(row["tweets"], stopwords)
        cleaned_sentences.append(cleaned)

    return cleaned_sentences


clean_sentences = get_cleaned_sentences(df, stopwords = True)
print(type(clean_sentences))
print("\n")
```

```python
txt_data = " "
i=0
for a in clean_sentences:
  txt_data = txt_data + ". " + clean_sentences[i]
  i+=1

print(txt_data)
```

```python
#tokenization
text = txt_data
sentence_list = nltk.sent_tokenize(text)
```

```python
print(sentence_list)
```

```python
def index_sort(list_var):
  length = len(list_var)
  list_index = list(range(0, length))

  x = list_var
  for i in range(length):
    for j in range(length):
      if x[list_index[i]] > x[list_index[j]]:
        temp = list_index[i]
        list_index[i] = list_index[j]
        list_index[j] = temp

  return list_index
```

```python
def response(user_ip):
  user_ip = user_ip.lower()
  sentence_list.append(user_ip)
  bot_response = ''
  cm = CountVectorizer().fit_transform(sentence_list)
  similarity_score = cosine_similarity(cm[-1], cm)
  similarity_score_list = similarity_score.flatten()
  index = index_sort(similarity_score_list)
  index = index[1:]
  response_flag = 0

  j=0
  for i in range(len(index)):
    if similarity_score_list[index[i]]>0.25:
      bot_response = bot_response+' '+sentence_list[index[i]]
      response_flag = 1
      j=j+1
    if j>2:
      break

  if response_flag ==0:
    bot_response = bot_response+' '+ "I dont know im sorry"
  
  sentence_list.remove(user_ip)
  
  return bot_response





  
```
